```python
def A():
    x = 10        # A의 지역 변수 x
    def B():
        nonlocal x    # 현재 함수의 바깥쪽에 있는 지역 변수 사용
        x = 20        # A의 지역 변수 x에 20 할당
 
    B()
    print(x)      # A의 지역 변수 x 출력
 
A()
```

```python
class Person:
    def __init__(self):
        self.__hello = '안녕하세요.' #__을 쓰면 private가 되고 이걸 사용하는 곳에서는 전부 __를 붙여야 함
 
    def greeting(self):
        print(self.__hello)
```

```python
def trace(func):                             # 호출할 함수를 매개변수로 받음
    def wrapper():                           # 호출할 함수를 감싸는 함수
        print(func.__name__, '함수 시작')    # __name__으로 함수 이름 출력
        func()                               # 매개변수로 받은 함수를 호출
        print(func.__name__, '함수 끝')
    return wrapper  
```

```python
def lunch():
    print("lunch")
```

```python
lunch_2 = trace(lunch)
lunch_2()
```

```python
@trace
def dinner():
    print("저녁")
dinner()
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.weather.go.kr/w/obs-climate/land/city-obs.do?auto_man=m&stn=0&dtm=&type=t99&reg=109&tm=2023.07.17.15%3A00"
res = requests.get(url)


soup = BeautifulSoup(res.text, 'html.parser')
table = soup.find('table',{'id':"weather_table"})
trs = table.find_all("tr")

trs_view = []
for tr in trs:
    tds = tr.find_all('td')
    if len(tds) == 0:
        continue
    mydata = [tds[0].find('a').text, tds[5].text, tds[9].text]
    trs_view.append(mydata)

trs_view
```
