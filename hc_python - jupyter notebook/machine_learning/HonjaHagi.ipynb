```python
wd = webdriver.Chrome()
store_list = []
number = 0
count = 0
for page in page_urls:
    count+=1
    print(count, end=" ")
    try:
        wd.get(page)
        time.sleep(2)
        html = wd.page_source
        soup = BeautifulSoup(html, 'html.parser')
        store_title = soup.select("h2.tit_location")[0].text
        store_score = soup.find('em',{'class':"num_rate"}).text.replace("점","")
        store_review = soup.find('strong',{'class':"total_evaluation"}).text.replace("전체","")
        if(float(store_score) >= 4.0):
            if(float(store_review.strip()) >= 20):
                number += 1
                store_list.append([number, store_title, store_score, store_review])
    except:
        continue
store_list
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
import time
import selenium

#원하는 검색지정
search_key = "부산대역 맛집"
print(search_key+ "을 검색하겠습니다.")

# 카카오 지도 오픈 후 부산대역 맛집 클릭
wd = webdriver.Chrome()
url = "https://map.kakao.com/"
wd.get(url)
searchbox=wd.find_element(By.ID,"search.keyword.query")
searchbox.send_keys(search_key)
searchbutton=wd.find_element(By.ID,"search.keyword.submit")
wd.execute_script("arguments[0].click();", searchbutton)
time.sleep(2)

print("150개 페이지의 정보를 담고 있습니다.")
# 검색 후 10페이지의 정보를 담음
allviews = []
count = 1

html = wd.page_source
soup = BeautifulSoup(html, "html.parser")
moreviews = soup.find_all(name="a", attrs={"class":"moreview"})
allviews.append(moreviews)
wd.find_element(By.ID,"info.search.place.more").send_keys(Keys.ENTER)
time.sleep(1)
for i in range(2,11):
    count+=1
    html = wd.page_source
    soup = BeautifulSoup(html, "html.parser")
    moreviews = soup.find_all(name="a", attrs={"class":"moreview"})
    allviews.append(moreviews)
    if i%5==0:
        wd.find_element(By.ID,"info.search.page.next").send_keys(Keys.ENTER)
        count-=5
        time.sleep(1)
    else:
        wd.find_element(By.ID,f"info.search.page.no{count+1}").send_keys(Keys.ENTER)
        time.sleep(1)

# 10페이지의 정보에서 주소를 담음
page_urls = []
for i in range(len(allviews)):
    for moreview in allviews[i]:
        page_url = moreview.get("href")
        page_urls.append(page_url)
wd.close()
page_urls, len(page_urls)

print("조건에 맞는 데이터를 찾고 있습니다.")
# 주소에 찾아가서 평점 4점 이상 리뷰 20개 이상인 데이터를 리스트에 담음
wd = webdriver.Chrome()
store_list = []
number = 0
count = 0
for page in page_urls:
    count+=1
    print(count, end=" ")
    try:
        wd.get(page)
        time.sleep(2)
        html = wd.page_source
        soup = BeautifulSoup(html, 'html.parser')
        store_title = soup.select("h2.tit_location")[0].text
        store_score = soup.find('em',{'class':"num_rate"}).text.replace("점","")
        store_review = soup.find('strong',{'class':"total_evaluation"}).text.replace("전체","")
        if(float(store_score) >= 4 and float(store_review.strip()) >= 2):
            number += 1
            store_list.append([number, store_title, store_score, store_review])
    except:
        continue
store_list
```
