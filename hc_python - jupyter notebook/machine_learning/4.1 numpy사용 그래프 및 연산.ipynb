```python
###########파이썬 제어와 numpy 제어 시간 비교 ####################################
#10000 x 1000     /  1000x 1  행렬계산
```

```python

n = 1000
m = 10000

X = np.random.rand(n,m)
W = np.random.rand(n,1)
Z = np.zeros((1,m))
```

```python
#행렬연산을 파이썬 제어
t1 = time.time()
for i in range(X.shape[1]):
    for j in range(X.shape[0]):
        Z[0][i] += W[j]*X[j][i]
t2 = time.time()

time_python = t2-t1
time_python #초단위
```

```python
# 같은 행렬 연산을 numpy로 수행
t1 = time.time()
Z = np.dot(W.T,X)
t2 = time.time()

time_numpy =  t2-t1
time_numpy
```

```python
# 파이썬은 이계산이 26초 걸렸고   넘파이 라이브러리 0.003초정도 걸린다 // 시간적 여유
```
