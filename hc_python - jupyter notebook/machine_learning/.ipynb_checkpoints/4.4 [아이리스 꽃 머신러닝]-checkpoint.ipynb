```python
#설치
#!pip install scikit-learn

#기본
import pandas as pd
import numpy as np

# iris 꽃의 데이터를 가져오자# 패키지 불러오기
from sklearn.datasets import load_iris 


# 데이터 값을 학습 및 피팅하는 DecisionTreeClassfier
from sklearn.tree import DecisionTreeClassifier


# seaborn의 데이터를 시각화하는 모듈을 가져오는 것이다.
import seaborn as sns

# 연습(학습)    test 직접 시험  구분
from sklearn.model_selection import train_test_split


from sklearn.metrics 
import f1_score, accuracy_score, confusion_matrix, precision_score, recall_score
```

```python
# iris 가 자진 데이터를 로드시켜보자
Iris = load_iris()
Iris.data
```

```python
# Iris의 각 column들이 가지는 의미들

#아이리스 여러가지 키중에서 feature_names이 각 컬럼들을 값을 의미
Iris.keys()
```

```python
# 컬럼들의 의미
Iris.feature_names
```

```python
# key 에 존재하는 target 의미하는 바 
# 0 --> 분류 이름'setosa' '
# 1  -> versicolor
# 2  -->virginica

Iris.target
```

```python
# 학습과 테스트의 비율을 나누는 것이다.
# iris.data = x( 꽃의 특성 데이터 값들), iris.target = y (꽃의 이름), test_size = 테스트가 전체의 차지하는 비율

X_train, X_test, y_train, y_test = train_test_split(
    Iris.data, Iris.target, test_size=0.2)
```

```python
# 나누어진 데이터들이 어떠한 형태의 모습을 가지고 있는지 보자.
# 트레인 세트가 120개, 테스트가 30개 X들은 4개의 컬럼씩 가지고 있고 Y들은 컬럼이 하나다.

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 데이터 학습 및 데이터 피팅
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
# 데이터베이스가 학습을 통해 얻은 자료를 통해 예측을 한다.

challenge = dt.predict(X_test)
challenge
```

```python
#  머신러닝이 예측한 값과 정답값을 비교해보자.

answer = y_test 
answer
```

```python
from sklearn.metrics import accuracy_score
y_pred = dt.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python

# 위의 값들을 비교해보면 3개의 값이 결과값과 예측값이 틀린것을 확인할 수 있다!

## 즉 30개중 27개의 답을 예측할 수 있었다.

## accuracy soore 를 통하여 어는정도 인지 알 수 있었다.
```

```python
#### 데이터 라벨링 ########
### 얻어진 테스트 값에 라벨을 붙이는 것이다.##


Xdf = pd.DataFrame(X_test.copy()) #테스트의 X_test 값을 카피해서 가져오고

Xdf["label"] = y_test             # 옆의 컬럼에 label을 만들어 y_test 값을 집어넣는다.
```

```python
# 컬럼의 이름을 붙이고
col = ['sepal length (cm)','sepal width (cm)', 'petal length (cm)', 'petal width (cm)', "label"]
Xdf.columns = col

# 30개의 완성된 X_test(데이터값들) 와 y_test(label) 표를 보자
Xdf
```

```python
# 맨위에 seqborn을 import 했고 이를 결과 값으로 보고싶어

# Xdf의 데이터를 pair plot(산점도 행렬)으로 시각화하는 코드입니다.
# hue="label"은 데이터의 특성 중 하나인 'label' 컬럼을 기준으로 데이터를 색상으로 구분하여 시각화하라는 의미입니다.

# label (0,1,2) 들이 색깔에 따라 분포도가 어떻게 되어있는지 확인하자.
# 산점도 어떻게 해석하는지 인터넷에 검색해서 보기

sns.pairplot(Xdf, hue="label")
```
